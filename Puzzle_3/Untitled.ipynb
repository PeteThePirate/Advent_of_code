```python
tree_map_numeric= [(j=='#') * 1 for i in open("./p3_input.txt").read().split("\n") for j in i ]

tree_map_numeric
```

```python
def move(tree_map, start_position, n_down, n_across):
        #start_position = (vertical_position, horizontal_position)
        
        dim_down, dim_across = tree_map.shape
        new_position =  (start_position[0] + n_down, (start_position[1] + n_across) % dim_across)
        
        return(new_position)
        
       
```

```python
#test 
print(tree_map_numeric.shape)
move(tree_map_numeric,(0,0), 5,35)
```

```python
def check_tree(tree_map, position):
    return(tree_map[position])
```

```python
check_tree(tree_map_numeric, (5,3))
```

```python
def move_to_end(tree_map, start_position, n_down, n_across):
    n_trees = 0
    current_position = start_position
    dim_down, dim_across = tree_map.shape
    
    while current_position[0] < dim_down-1:
        current_position = move(tree_map, current_position, n_down, n_across)
        n_trees += check_tree(tree_map, current_position)
        
    return([n_trees, current_position])
```

Check the slope 1-down, 3 across

```python
move_to_end(tree_map_numeric, (0,0), 1, 3)
```

Check multiple slopes and multiply their number of trees

```python
slopes_to_check= [(1,1),(1,3),(1,5),(1,7),(2,1)]

results = [move_to_end(tree_map_numeric, (0,0), i[0], i[1]) for i in slopes_to_check]
```

```python
results
```

```python
tree_counts=[i[0] for i in results]

tree_counts
```

```python
np.prod(tree_counts) #seems to give an integer overflow problem

answer = 1
for i in tree_counts:
    answer = answer * int(i)
    
answer
```
